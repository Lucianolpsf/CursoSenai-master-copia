```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rc('figure',figsize = (5,5))
figure, axes = plt.subplots()
```

```python
df = pd.read_csv("housing.csv")
```

```python
df.head(3)
```

```python
bedroom_counts = df.bedrooms.value_counts()
```

```python
bedroom_counts
```

```python
area = plt.figure()
g1 = area.add_subplot(1,1,1)
```

```python
g1.barh(bedroom_counts.index, bedroom_counts.values, color='skyblue')

# Inverte os rótulos e título
g1.set_ylabel('Quantidade de Quartos')
g1.set_xlabel('Número de Casas')
g1.set_title('Quantidade de Casas por Número de Quartos')
area
```

```python
area2 = plt.figure()
g2 = area2.add_subplot(1,1,1)
g2.bar(bedroom_counts.index, bedroom_counts.values, color='skyblue')
```

```python
g2.set_ylabel('Quantidade de Casas')
g2.set_xlabel('Número de Quartos')
g2.set_title('Quantidade de Casas por Número de Quartos')
area2
```

```python
# Adiciona bordas e cor de fundo
g2.spines['top'].set_visible(False)
g2.spines['right'].set_visible(False)
g2.spines['bottom'].set_color('gray')
g2.spines['left'].set_color('gray')

# Adiciona uma grade
g2.grid(True, linestyle='--', alpha=0.7)

# Adiciona rótulos e título
g2.set_ylabel('Quantidade de Casas')
g2.set_xlabel('Número de Quartos')
g2.set_title('Quantidade de Casas por Número de Quartos')

# Adiciona cor de fundo ao gráfico
g2.set_facecolor('#F9F9F9')
```

```python
area3 = plt.figure()
g3 = area3.add_subplot(1, 1, 1)

# Exemplo com a variável 'price'
g3.scatter(df['price'], df['area'])

g3.set_xlabel('Área')
g3.set_ylabel('Preço')
g3.set_title('Preço x Área')
```

```python
area4 = plt.figure()
g4 = area4.add_subplot(1, 1, 1)

# Plotando o gráfico de pizza
g4.pie(bedroom_counts, labels=bedroom_counts.index, autopct='%1.1f%%', startangle=90)

# Construindo a legenda com rótulos e quantidades
legend_labels = [f'{label} ({count})' for label, count in zip(bedroom_counts.index, bedroom_counts)]

# Adicionando legenda
g4.legend(legend_labels, title='Número de Quartos', loc='upper right', bbox_to_anchor=(1, 0.9))

# Adicionando título
g4.set_title('Distribuição de Quartos')
```
